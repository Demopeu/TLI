```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "NFLX.csv"
df = pd.read_csv(csv_path)
df['Date'] = pd.to_datetime(df['Date'])
B = df['Date'].dt.year >= 2021
re_df = df[B]
new_df = re_df.groupby(re_df['Date'].dt.to_period("M")).mean()
```

```python
plt.plot(new_df['Date'], new_df['Close'])

plt.title('Monthly Average Close Prices')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Average Close Price')

# 그래프 출력
plt.show()
```
