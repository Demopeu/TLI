```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = "NFLX.csv"
df = pd.read_csv(csv_path)
df['Date'] = pd.to_datetime(df['Date'])
B = df['Date'].dt.year >= 2022
new_df = df[B]
```

```python
plt.plot(new_df['Date'], new_df['Close'], label='Close')
plt.plot(new_df['Date'], new_df['Low'], label='Low')
plt.plot(new_df['Date'], new_df['High'], label='High')

plt.title('High, Low, and Prices since January 2022')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.ylabel('Price')

# 범례 표시
plt.legend()

# 그래프 표시
plt.show()
```
